## Demo: How to work with database


Download database file.


```python
import urllib.request

# url to download
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m4_survey_data.sqlite'

# Path to save the file
file_path = 'm4_survey_data.sqlite'

# Download the file
urllib.request.urlretrieve(url, file_path)
```

Connect to the database.


```python
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

Import pandas module.


```python
import pandas as pd
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


```python
# your code goes here
query = """ SELECT * 
            FROM master;"""
df = pd.read_sql(query, conn)
```

```python
df.set_index('index', inplace=True)
```

```python
pd.set_option("display.max_columns", None)
df.head()
```

## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(data=df, x='ConvertedComp')
```

### Box Plots


Plot a box plot of `Age.`


```python
q1 = df.Age.describe()['25%']
q3 = df.Age.describe()['75%']
iqr = q3 - q1
upper_lim = q3 + (iqr * 1.5)
lower_lim = q1 - (iqr * 1.5)
```

```python
sns.boxplot(data=df, x='Age')
plt.axvline(x=upper_lim, color='r', linestyle='--', label='Upper Limit')
plt.axvline(x=lower_lim, color='b', linestyle='--', label='Lower Limit')
plt.title('Age Boxplot')
plt.legend()
plt.show()
```

### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
sns.scatterplot(data=df, x='Age', y='WorkWeekHrs')
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
sns.scatterplot(data=df, x='WorkWeekHrs', y='CodeRevHrs', size='Age', sizes=(20, 200))
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
query = """SELECT * FROM DatabaseDesireNextYear;"""
df_database = pd.read_sql(query, conn)
filtered_df = df_database.DatabaseDesireNextYear.value_counts().head()
plt.pie(filtered_df, labels=filtered_df.index, autopct='%1.1f%%')
plt.title('Top 5 databases for the next Year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
import seaborn as sns
import matplotlib.pyplot as plt

# Filter the DataFrame for the age group 30 to 35
filtered_df = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

# Calculate the median values for WorkWeekHrs and CodeRevHrs
medians = filtered_df.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median()

# Create the stacked bar chart
medians.plot(kind='bar', stacked=True)

# Set plot title and labels
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.xlabel('Age')
plt.ylabel('Median Hours')

# Show the plot
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
filtered_df = df[(df.Age >= 45) & (df.Age <= 60)]
medians = filtered_df.groupby('Age')['ConvertedComp'].median()
sns.lineplot(data=medians)
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
sns.countplot(y='MainBranch', data=df)

# Set plot title and labels
plt.title('Main Branch Distribution')
plt.xlabel('Count')
plt.ylabel('Main Branch')

# Show the plot
plt.show()
```
